```python
from cvzone.HandTrackingModule import HandDetector
import cv2
import os
import time
import webbrowser

def perform_action(action):
    if "http" in action:
        webbrowser.open(action)
        time.sleep(2)
    else:
        os.system(action)
        time.sleep(2)

menu_options = {
    "open_youtube": [
        [0, 1, 0, 0, 0],  # Index finger up
        [
            ("Open Playlist 1", "https://youtu.be/yUB1_sfaGqg"),
            ("Open Playlist 2", "https://youtu.be/QN0BfamqA4Y"),
        ]
    ],
    "open_apps": [
        [0, 1, 1, 0, 0],  # Two finger up
        [
            ("Open Calculator", "calc"),
            ("Open Notepad", "notepad"),
        ]
    ],
    "open_websites": [
        [0, 1, 1, 1, 0],  # Three finger up
        [
            ("Open Google", "https://www.google.com"),
            ("Open GitHub", "https://www.github.com"),
        ]
    ],
    # Add more menu options here
}

model = HandDetector()
cap = cv2.VideoCapture(0)

while True:
    status, photo = cap.read()
    cv2.imshow("Menu", photo)

    if cv2.waitKey(100) == 13:
        break

    hands, photo = model.findHands(photo, draw=True)

    if hands:
        hand = hands[0]
        fingeruplist = model.fingersUp(hand)

        for option, (gesture, actions) in menu_options.items():
            if fingeruplist == gesture:
                print(f"Selecting option: {option}")
                for action_name, action in actions:
                    print(f"Performing action: {action_name}")
                    perform_action(action)
                break  # Exit the loop after opening one option

cv2.destroyAllWindows()
```
